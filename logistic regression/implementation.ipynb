### Logistics Regression Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
## create the dataset
X , y = make_classification(n_classes=2, n_samples=1000, n_features=10, random_state=43)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train , y_test = train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(X_train, y_train)
```

```python
y_pred = log.predict(X_test)
```

```python
y_pred
```

```python
### Performance Metrics
from sklearn.metrics import accuracy_score, confusion_matrix , classification_report
```

```python
score = accuracy_score(y_test, y_pred)
score
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print(classification_report(y_test, y_pred))
```

### Hyper Parameter Tuning and Cross Validation

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params = dict(penalty = penalty , C = c_values , solver=solver)
```

```python
params
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator= model,param_grid= params ,scoring= 'accuracy' , cv=cv)
```

```python
grid
```

```python
grid.fit(X_train , y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
score = accuracy_score(y_test , y_pred)
score
```

### Randomized Search CV

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import RandomizedSearchCV
random = RandomizedSearchCV(estimator=model, param_distributions=params, cv = cv ,scoring='accuracy' )
```

```python
random.fit(X_train , y_train)
```

```python
random.best_params_
```

```python
random.best_score_
```

```python
y_pred = random.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
score = accuracy_score(y_test , y_pred)
score
```
