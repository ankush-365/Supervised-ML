```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Student_Performance.csv')
df

df['Extracurricular_Activities'] = df['Extracurricular_Activities'].map({'Yes':1,'No':0})
```

```python
sns.heatmap(df.corr(),annot=True,cmap='RdYlGn')
```

```python
sns.boxplot(df['Hours Studied'])
```

```python
y=df.pop("Performance Index")
x=df
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(df.columns)
print(lr.coef_)
```

```python
result_train=lr.predict(x_train)
result_test=lr.predict(x_test)
```

```python
acc=r2_score(y_train,result_train)
acc
```

```python
acc_test=r2_score(y_test,result_test)
acc_test

```

```python
plt.scatter(x["Hours Studied"],y)
```

```python
plt.scatter(y_test,result_test)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='k',ls='--')
```
