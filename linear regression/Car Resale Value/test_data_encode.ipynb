### Encoding Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lblencoder = LabelEncoder()
```

```python
df['model'] = lblencoder.fit_transform(df['model'])
```

```python
df.head()
```

```python
df['fuel_type'] = lblencoder.fit_transform(df['fuel_type'])
df['brand'] = lblencoder.fit_transform(df['brand'])
```

```python
df.head()
```

```python
df['engine'] = lblencoder.fit_transform(df['engine'])
```

```python
df['transmission'] = lblencoder.fit_transform(df['transmission'])
```

```python
def simplify_color(color):
    color = str(color).lower()
    if 'black' in color:
        return 'black'
    elif 'white' in color or 'snow' in color or 'pearl' in color:
        return 'white'
    elif 'gray' in color or 'silver' in color or 'platinum' in color:
        return 'gray/silver'
    elif 'red' in color or 'maroon' in color or 'crimson' in color:
        return 'red'
    elif 'blue' in color or 'navy' in color:
        return 'blue'
    elif 'green' in color:
        return 'green'
    elif 'brown' in color or 'bronze' in color or 'tan' in color:
        return 'brown'
    elif 'yellow' in color or 'gold' in color:
        return 'yellow/gold'
    elif 'orange' in color:
        return 'orange'
    elif 'purple' in color or 'violet' in color:
        return 'purple'
    else:
        return 'other'
```

```python
df['ext_col'] = df['ext_col'].apply(simplify_color)
df['int_col'] = df['int_col'].apply(simplify_color)
```

```python
df.head()
```

```python
df['ext_col'] = lblencoder.fit_transform(df['ext_col'])
df['int_col'] = lblencoder.fit_transform(df['int_col'])
```

```python
df['accident'] = df['accident'].map({'None reported':0, 'At least 1 accident or damage reported': 1 ,np.nan:0  })
```

```python
df['clean_title'] = df['clean_title'].map({'Yes':1 , np.nan:0})
```

```python
df.head()
```

```python
df.to_csv('test_data_encoded.csv')
```
