```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=1000, n_classes=2, n_features=2, n_clusters_per_class=2, n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(pd.DataFrame(X))
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train ,y_test = train_test_split(pd.DataFrame(X),y,test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
y_train
```

```python
svc.fit(X_train,y_train)
```

```python
## Predictions
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
score = accuracy_score(y_test,y_pred)
score
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
rbf = SVC(kernel='rbf',gamma=1,C=100)
```

```python
rbf.fit(X_train,y_train)
```

```python
y_pred1 = rbf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
score = accuracy_score(y_test,y_pred1)
score
```

```python
## Params
from sklearn.model_selection import GridSearchCV

params = {
    'C': [0.1,1,10,100,1000],
    'gamma': [1,0.1,0.01,0.001],
    'kernel': ['linear','sigmoid','rbf','poly']
}
```

```python
grid = GridSearchCV(SVC(),param_grid=params,refit=True, cv=5, verbose=3 )
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_index_
```
