```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
diabetes
```

```python
print(diabetes['DESCR'])
```

```python
## Dependent and Independent Feature
X = pd.DataFrame(diabetes['data'], columns=['age', 'sex', 'BMI', 'BP', 'S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
y = diabetes['target']
```

```python
X
```

```python
y
```

```python
### Train test split
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X,y, test_size=0.2, random_state=10)
```

```python
X_train.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(X_train, y_train)
```

```python
y_train_pred = dtr.predict(X_train)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_train, y_train_pred)) ## Overfitting
```

```python
y_pred = dtr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

### Hyper Parameter Tuning

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

```python
params = {
    'criterion' : ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5,10,15,20],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(dtr , param_grid=params, cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_t_pred = grid.predict(X_train)
from sklearn.metrics import r2_score
print(r2_score(y_train, y_t_pred)) ##Training Score
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred)) ## Testing Score
```
