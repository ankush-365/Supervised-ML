```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print(iris['DESCR'])
```

```python
iris['data']
```

```python
## Independent feature
X = pd.DataFrame(iris['data'],columns=['sepal length', 'sepal width','petal length', 'petal width'])
```

```python
## Dependent Feature
y = iris['target']
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

### Decision Tree: Standardization is not necessary

```python
## Model Training
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier(max_depth=2) ## Post Pruning
```

```python
treeClassifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier , filled=True)
```

```python
y_train_pred = treeClassifier.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, y_train_pred)) ## Training Score
```

```python
y_pred = treeClassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f" Testing Score : {accuracy_score(y_test, y_pred)}") ## Testing Score
```

```python
## Model Training
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier(max_depth=2) ## Post Pruning
```

## Decision Tree with Pre-Pruning and Hyper Parameter Tuning

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5],
    'max_features': ['auto', 'sqrt', 'log2'] 
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(treeClassifier, param_grid=params, cv=5 , scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_test, y_pred))
```
